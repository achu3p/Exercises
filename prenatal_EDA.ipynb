```sas
/*
MAGER - maternal age 12-43  | maternal characteristic
MEDUC - 1 8th/less / 2 9-12 no diploma / 3 HS/GED / 4 some college / 5 associate's / 6 bachelor's / 7 master's / 8 doctorate/professional  | maternal characteristic
BMI - 13-69.9 | maternal health
PWgt_R - 075-375 prepregnancy weight | maternal health
firstbirthbin - 1 experienced / 0 first birth | maternal characteristic
cigbin - 1 smoked / 0 never smoked | maternal health
riskfactorbin - 1 risk factor present / 0 no risk factors | maternal health
infectionbin - 1 infection present / 0 no infection | maternal health
adeqcare - adequate / otherwise 
adeqcarebin - 0 adequate / 1 otherwise | RESPONSE VARIABLE - GETTING ADEQUATE PRENATAL CARE
ILIVE - infant living; compare with adequacy of prenatal care (ilive as response) to show importance of prenatal care
*/

/*set new variables for analysis*/
data nat3;
    set nat2;
    if previs_rec=01 or previs_rec=02 or previs_rec=03 or previs_rec=04 then PrenatalVisits="6/Less Visits";
    if previs_rec=05 or previs_rec=06 or previs_rec=07 or previs_rec=08 or previs_rec=09 or previs_rec=10 or previs_rec=11 then PrenatalVisits="7+ Visits";
    if precare5=1 and prenatalvisits="7+ Visits" then adeqcare="adequate";
    if precare5=2 and prenatalvisits="7+ Visits" then adeqcare="adequate";
    if precare5=3 or precare5=4 or prenatalvisits="6/Less Visits" then adeqcare="otherwise";
    if adeqcare="adequate" then adeqcarebin=0;
        else adeqcarebin=1;
    if priorlive>0 or priordead>0 then firstbirthbin=1;
        else firstbirthbin=0;
    if CIG0_R>0 or CIG1_R>0 or CIG2_R>0 or CIG3_R>0 then cigbin=1;
        else cigbin=0;
    if RF_PDIAB='Y' or RF_GDIAB='Y' or RF_PHYPE='Y' or RF_GHYPE='Y' or RF_EHYPE='Y' or RF_PPTERM='Y' or RF_FEDRG='Y' or RF_CESAR='Y' then riskfactorbin=1;
        else riskfactorbin=0;
    if IP_GON="Y" or IP_SYPH="Y" or IP_HEPB="Y" or IP_HEPC="Y" or IP_CHLAM="Y" then infectionbin=1;
        else infectionbin=0;
    drop previs_rec precare5 prenatalvisits priorlive priordead CIG0_R CIG1_R CIG2_R CIG3_R RF_PDIAB RF_GDIAB RF_PHYPE RF_GHYPE RF_EHYPE RF_PPTERM RF_FEDRG RF_ARTEC RF_CESAR IP_GON IP_SYPH IP_HEPB IP_HEPC IP_CHLAM;
run;
```

```sas
/*why adequate prenatal care is important!*/

proc sort data=nat3;
    by ilive adeqcare;
run;

proc freq data=nat3 order=freq noprint;
    by ilive; 
    tables adeqcare / out=iliveadeqcarefreq;
run;

proc sgplot data=iliveadeqcarefreq noborder;
    title1 justify=left color=lightslategray "Infant Mortality at Time of Report & Adequacy of Prenatal Care";
    hbar ilive / response=Percent group=adeqcare
                     grouporder=data groupdisplay=stack
                     barwidth=.6 seglabel;
    yaxis label="Living" labelattrs=(color=lightslategray) display=(noticks) valueattrs=(color=lightslategray) labelpos=top;
    xaxis grid values=(0 to 100 by 10) label="Percentage" labelattrs=(color=darkgray) display=(noline noticks) valueattrs=(color=darkgray) labelpos=left;
    keylegend / position=bottomleft titleattrs=(color=lightslategray) valueattrs=(color=lightslategray);
    label adeqcare="Prenatal Care Adequacy";
run;
title1;

/*
95% of infants who lived at time of report had mothers who received adequate prenatal care, vs 37% for not living
alternatively 
sets up that prenatal care is correlated w/ mortality THUS PRENATAL CARE IS IMPORTANT (so what might make a mother more likely to come in & check?); RUN CORRELATION TEST
*/
```

```sas
/*
MAGER - maternal age 12-43  | maternal characteristic
MEDUC - 1 8th/less / 2 9-12 no diploma / 3 HS/GED / 4 some college / 5 associate's / 6 bachelor's / 7 master's / 8 doctorate/professional  | maternal characteristic
BMI - 13-69.9 | maternal health
PWgt_R - 075-375 prepregnancy weight | maternal health
firstbirthbin - 1 experienced / 0 first birth | maternal characteristic
cigbin - 1 smoked / 0 never smoked | maternal health
riskfactorbin - 1 risk factor present / 0 no risk factors | maternal health
infectionbin - 1 infection present / 0 no infection | maternal health
adeqcare - adequate / otherwise 
adeqcarebin - 0 adequate / 1 otherwise | RESPONSE VARIABLE - GETTING ADEQUATE PRENATAL CARE
ILIVE - infant living; compare with adequacy of prenatal care (ilive as response) to show importance of prenatal care
*/

proc logistic data=nat3;
    class firstbirthbin cigbin riskfactorbin infectionbin / param=ref ref=first;
    model adeqcarebin(event="0") = mager meduc bmi pwgt_r firstbirthbin cigbin riskfactorbin infectionbin;
    ods select parameterestimates;
run;
```

```sas
proc logistic data=nat3;
    class firstbirthbin cigbin riskfactorbin infectionbin / param=ref ref=first;
    model adeqcarebin(event="0") = mager meduc bmi pwgt_r firstbirthbin cigbin riskfactorbin infectionbin / rsquare;
    ods select association rsquare;
run;
proc logistic data=nat3 plots(only)=roc;
    class firstbirthbin cigbin riskfactorbin infectionbin / param=ref ref=first;
    model adeqcarebin(event="0") = mager meduc bmi pwgt_r firstbirthbin cigbin riskfactorbin infectionbin;
run;
/*AUC ~ .56, barely better than random chance */
```
